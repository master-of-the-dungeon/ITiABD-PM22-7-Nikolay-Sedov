```python
import sympy as sp
```

```python
A_sympy = sp.Matrix([[0.04, 0.86], [0.89, 0.08]])
Y_sympy = sp.Matrix([540, 581])
I_sympy = sp.eye(2)
```

```python
X_sympy = (I_sympy - A_sympy).inv() * Y_sympy
X_sympy
```

```python
_ = sum(X_sympy)
```

```python
Z_sympy = A_sympy * X_sympy
```
