```python
import pandas as pd

# Load the provided Excel file
file_path = 'dataa.xlsx'
data = pd.read_excel(file_path)

# Display the first few rows of the dataframe to understand its structure
data.head()
```

```python

data_new = pd.read_excel(file_path, engine='openpyxl')

data_new.head()
```

```python
import statsmodels.formula.api as smf

formula = 'Y ~ X1 + X2 + X3 + X4 + X5 + X6'


model = smf.ols(formula, data=data_new).fit()
model_summary = model.summary()
model_summary
```

# FE model

```python
from linearmodels.panel import PanelOLS
from linearmodels.panel import compare

data_new = data_new.set_index(['Reg', 'Year'])

fe_model = PanelOLS.from_formula('Y ~ X1 + X2 + X3 + X4 + X5 + X6 + EntityEffects', data=data_new)
fe_results = fe_model.fit()

fe_results
```
