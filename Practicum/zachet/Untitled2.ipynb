Создать компьютерную игру «Охотник» (поле 600х600). В качестве 
спрайтов использовать изображения охотника и зайца. Охотник 
должен стрелять по нажатию на «пробел». Заяц должен двигаться (при 
нажатии соответствующих кнопок на клавиатуре). Конец игры при 
попадании выстрелом по зайцу.

```python
import tkinter as tk
import random

def shoot(event):
    global hunter, rabbit
    x1, y1, x2, y2 = canvas.bbox(hunter)
    overlaps = canvas.find_overlapping(x1, y1, x2, y2)
    if rabbit in overlaps:
        canvas.delete(rabbit)
        canvas.create_text(300, 300, text="Вы поймали зайца!", font=("Arial", 24))

def move_rabbit():
    global rabbit
    x = random.randint(50, 550)
    y = random.randint(50, 550)
    canvas.coords(rabbit, x, y, x+50, y+50)
    canvas.after(1000, move_rabbit)

root = tk.Tk()
root.title("Охотник")

canvas = tk.Canvas(root, width=600, height=600, bg='lightgreen')
canvas.pack()

hunter_img = tk.PhotoImage(file="hunter.png")
hunter = canvas.create_image(50, 50, anchor=tk.NW, image=hunter_img)

rabbit_img = tk.PhotoImage(file="rabbit.png")
rabbit = canvas.create_image(300, 300, anchor=tk.NW, image=rabbit_img)

root.bind("<space>", shoot)
move_rabbit()

root.mainloop()
```

 Создать компьютерную игру «Собери фрукты». Рандомно на поле 
появляются фрукты и исчезают через n-секунд. Баллы начисляются при 
клике мышью по фрукту.


```python
import tkinter as tk
import random

def create_fruit():
    x = random.randint(50, 550)
    y = random.randint(50, 550)
    fruit_type = random.choice(fruit_images)
    fruit = canvas.create_image(x, y, anchor=tk.CENTER, image=fruit_type)
    fruits.append(fruit)
    canvas.after(3000, delete_fruit, fruit)

def delete_fruit(fruit):
    canvas.delete(fruit)
    fruits.remove(fruit)

def collect(event):
    global score
    x, y = event.x, event.y
    for fruit in fruits:
        x1, y1, x2, y2 = canvas.coords(fruit)
        if x1 < x < x2 and y1 < y < y2:
            canvas.delete(fruit)
            fruits.remove(fruit)
            score += 1
            score_label.config(text=f"Score: {score}")
            break

root = tk.Tk()
root.title("Собери фрукты")

canvas = tk.Canvas(root, width=600, height=600, bg='lightgreen')
canvas.pack()

apple_img = tk.PhotoImage(file="apple.png")
banana_img = tk.PhotoImage(file="banana.png")
orange_img = tk.PhotoImage(file="orange.png")

fruit_images = [apple_img, banana_img, orange_img]
fruits = []
score = 0

score_label = tk.Label(root, text="Score: 0", font=("Arial", 18))
score_label.pack()

canvas.bind("<Button-1>", collect)

root.after(1000, create_fruit)  # Создание фруктов каждую секунду

root.mainloop()
```

Создать компьютерную игру «Собери фрукты2». Рандомно на поле 
появляются фрукты и овощи и исчезают через n-секунд. Баллы

начисляется.


```python
import tkinter as tk
import random

def create_item():
    x = random.randint(50, 550)
    y = random.randint(50, 550)
    item_type = random.choice(item_images)
    item = canvas.create_image(x, y, anchor=tk.CENTER, image=item_type)
    items.append((item, item_type))
    canvas.after(3000, delete_item, item)

def delete_item(item):
    canvas.delete(item)
    for i, data in enumerate(items):
        if data[0] == item:
            items.pop(i)
            break

def collect(event):
    global score
    x, y = event.x, event.y
    for item, item_type in items:
        x1, y1, x2, y2 = canvas.coords(item)
        if x1 < x < x2 and y1 < y < y2:
            if item_type in fruit_images:
                score += 1
                score_label.config(text=f"Score: {score}")
            canvas.delete(item)
            items.remove((item, item_type))
            break

root = tk.Tk()
root.title("Собери фрукты2")

canvas = tk.Canvas(root, width=600, height=600, bg='lightgreen')
canvas.pack()

apple_img = tk.PhotoImage(file="apple.png")
banana_img = tk.PhotoImage(file="banana.png")
orange_img = tk.PhotoImage(file="orange.png")
carrot_img = tk.PhotoImage(file="carrot.png")
tomato_img = tk.PhotoImage(file="tomato.png")

fruit_images = [apple_img, banana_img, orange_img]
item_images = [apple_img, banana_img, orange_img, carrot_img, tomato_img]

items = []
score = 0

score_label = tk.Label(root, text="Score: 0", font=("Arial", 18))
score_label.pack()

canvas.bind("<Button-1>", collect)

root.after(1000, create_item)  # Создание предметов каждую секунду

root.mainloop()
```

Создать компьютерную игру «Собери монеты». Рандомно на поле 
появляются монеты (1 рубль, 2 рубля и 5 рублей) и исчезают через n-
секунд. Баллы начисляются при клике мышью по монетам. За монету 
достоинством в 1 рубль дается 1 балл. За монету достоинством в2 
рубля дается 2 балла. За монету достоинством в 5 рублей дается 5 
баллов.

```python
import tkinter as tk
import random

def create_coin():
    x = random.randint(50, 550)
    y = random.randint(50, 550)
    coin_type = random.choice(coin_images)
    coin = canvas.create_image(x, y, anchor=tk.CENTER, image=coin_type)
    coins.append((coin, coin_type))
    canvas.after(3000, delete_coin, coin)

def delete_coin(coin):
    canvas.delete(coin)
    for i, data in enumerate(coins):
        if data[0] == coin:
            coins.pop(i)
            break

def collect(event):
    global score
    x, y = event.x, event.y
    for coin, coin_type in coins:
        x1, y1, x2, y2 = canvas.coords(coin)
        if x1 < x < x2 and y1 < y < y2:
            if coin_type == one_coin:
                score += 1
            elif coin_type == two_coins:
                score += 2
            elif coin_type == five_coins:
                score += 5
            score_label.config(text=f"Score: {score}")
            canvas.delete(coin)
            coins.remove((coin, coin_type))
            break

root = tk.Tk()
root.title("Собери монеты")

canvas = tk.Canvas(root, width=600, height=600, bg='lightblue')
canvas.pack()

one_coin_img = tk.PhotoImage(file="one_coin.png")
two_coins_img = tk.PhotoImage(file="two_coins.png")
five_coins_img = tk.PhotoImage(file="five_coins.png")

coin_images = [one_coin_img, two_coins_img, five_coins_img]

coins = []
score = 0

score_label = tk.Label(root, text="Score: 0", font=("Arial", 18))
score_label.pack()

one_coin = one_coin_img
two_coins = two_coins_img
five_coins = five_coins_img

canvas.bind("<Button-1>", collect)

root.after(1000, create_coin)  # Создание монет каждую секунду

root.mainloop()
```
