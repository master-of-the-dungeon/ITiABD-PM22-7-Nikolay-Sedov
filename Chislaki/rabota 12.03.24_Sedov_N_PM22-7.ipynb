```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def f(x):
    return x*np.exp(x)


a = 0
b = 1
n = 6 # n = 6,   ([0. , 0.2, 0.4, 0.6, 0.8, 1. ]),  h = (a-b)/2   0.1 

def simp(a,b,n):
    z = np.linspace(a,b,n+1)
    I = 0
    for i in range(len(z)-1):
        I += ((z[i+1] - z[i])/6)*(f(z[i])+f(z[i+1])+4*f((a+b)/2))
    return I

simp(a,b,n)
```

```python
def solve(f, e=0.00005,left=-1000,right=1000):
    default = []
    for i in np.arange(left,right,0.1):
        if np.sign(f(i)) + np.sign(f(i-1)) == 0:
            default.append((i-1,i))
            
    sols = []
    
    for i in default:
        ss = []
        z = 1000
        while abs(z) > e:
            ss = [f(j) for j in i]
            for k in range(1,len(ss)):
                 if np.sign(ss[k])+np.sign(ss[k-1]) == 0:
                        i = [i[k-1],(i[k]+i[k-1])/2,i[k]]
                        z = ss[k]+ss[k-1]
        try:
            if abs(((i[k]+i[k-1])/2) - sols[-1])>e and abs(f((i[k]+i[k-1])/2))<e:
                sols.append((i[k]+i[k-1])/2)
        except:
            sols.append((i[k]+i[k-1])/2)
        
    return sols
```

```python
def f1(x):
    return 3 * np.sin(x) + x - 2

solve(f1,e=10**(-3))
```

```python
def seidel(a, x ,b):
    n = len(a)
    for j in range(0, n):
        d = b[j]
        for i in range(0, n):
            if(j != i):
                d-=a[j][i] * x[i]
        x[j] = d / a[j][j]
    return x 

a = [[1,0.96,0.78], [1.06, 1, 1.1], [1.15, 1.1, 1]]
b = [1, 1.27, 0.82]
x = [0,0,0]

for i in range(100): 
    x = seidel(a, x, b)
print(x)
```
