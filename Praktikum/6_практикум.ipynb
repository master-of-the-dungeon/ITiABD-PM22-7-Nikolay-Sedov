```python
import matplotlib.pyplot as plt
from sympy import *
import numpy as np
import math
from itertools import islice, count
```

Реализовать функцию визуализации последовательности полигонов, представленной в виде итератора 

```python
#1

n = int(input('Количество фигур '))
x=[]
y = []
for i in islice(count(1), n):
    x.append(i)
    y.append(0)

plt.plot(x, y, linestyle='', marker='o', markersize=20)
plt.grid() 
plt.title('Circle')
plt.xlabel("Ось Х")
plt.ylabel("Ось Y")
plt.show() 
```

Реализовать функции, генерирующие бесконечную последовательность не пересекающихся полигонов с различающимися координатами (например, «ленту»): 

прямоугольников ( gen_rectangle() ); 

треугольников ( gen_triangle() ); 

правильных шестиугольников ( gen_hexagon() ). 

с помощью данных функций используя функции из модуля itertools сгенерировать 7 фигур, включающих как прямоугольники, так и треугольники и шестиугольники, визуализировать результат. 

```python
#2.1
def gen_rectangle(): 
  for i in islice(count(0), n):    
    x = [i*shag, sh+i*shag, sh+i*shag, i*shag, i*shag] 
    y =  [1,1,1+vs,1+vs,1]   
    plt.plot(x,y)

shag = int(input('Введите расстояние между фигурами (шаг) '))
n = int(input('Введите колиечество фигур '))
sh = int(input('Ширина '))
vs = int(input('Высота '))

gen_rectangle() 
plt.title('Rectangle')
plt.grid()
plt.xlabel("Ось Х") 
plt.ylabel("Ось Y") 
plt.show() 
```

```python
#2.2
def gen_triangle():
  for i in islice(count(0), n):
    x = [i*shag, sh//2 + i*shag, sh + i*shag , i*shag]
    y = [1, 1+vs,1, 1]
    plt.plot(x,y)
shag = int(input('Введите расстояние между фигурами (шаг) '))
n = int(input('Введите колиечество фигур '))
sh = int(input('Ширина '))
vs = int(input('Высота '))


gen_triangle() 
plt.grid() 
plt.title('Triangle')
plt.xlabel("Ось Х") 
plt.ylabel("Ось Y") 
plt.show() 
```

```python
#2.3
def gen_hexagon():
  for i in islice(count(0), n):
    x = [1 + sh//2 + i*shag, 1 + i*shag, 1+i*shag, 1 + sh//2 + i*shag, 1 + sh + i*shag, 1 + sh + i*shag, 1 + sh//2 + i*shag] 
    y = [1,1 + vs//3, 1 + 2*vs//3, 1 + vs, 1 + 2*vs//3, 1 + vs//3, 1]
    plt.plot(x,y)
shag = int(input('Введите расстояние между фигурами (шаг) '))
n = int(input('Введите количество фигур '))
sh = int(input('Ширина '))
vs = int(input('Высота '))

gen_hexagon() 
plt.grid() 
plt.title('Hexagon')
plt.xlabel("Ось Х") 
plt.ylabel("Ось Y") 
plt.show() 
```

Реализовать операции:  

параллельный перенос ( tr_translate ); 

поворот ( tr_rotate ); 

симметрия ( tr_symmetry );  

гомотетия ( tr_homothety ); 

```python
#3.1
def tr_translate():
  for i in islice(count(0), n):
    x = [i*shag, sh+i*shag, sh+i*shag, i*shag, i*shag] 
    y =  [1,1,1+vs,1+vs,1]   
    plt.plot(x,y)
    plt.plot(x,y)
    for k in islice(count(0), 3):
      x = [i*shag, sh+i*shag, sh+i*shag, i*shag, i*shag] 
      y =  [1+k*shag, 1+k*shag, 1 + vs + k*shag, 1 + vs + k*shag, 1+k*shag]  
      plt.plot(x,y)

shag = int(input('Введите расстояние между фигурами (шаг) '))
n = int(input('Введите количество фигур '))
sh = int(input('Ширина '))
vs = int(input('Высота '))

tr_translate()
plt.grid() 
plt.title('Rectangle')
plt.xlabel("Ось Х") 
plt.ylabel("Ось Y") 
# plt.axis('square') 
plt.show() 
```

```python
#3.3 и 4.3
def tr_symmetry():
  for i in islice(count(0), n):
    x1 = [i*shag,sh//2 + i*shag, sh + i*shag , i*shag]
    y1 = [1, 1+vs,1, 1]
    plt.plot(x1,y1)
    for k in islice(count(1), 1):
      x = [i*shag, sh//2 + i*shag, sh + i*shag, i*shag] 
      y = [1 + vs + k*shag, 1+k*shag, 1 + vs + k*shag, 1 + vs + k*shag]  
      plt.plot(x,y)

shag = int(input('Введите расстояние между фигурами (шаг) '))
n = int(input('Введите колиечество фигур '))
sh = int(input('Ширина '))
vs = int(input('Высота '))

tr_symmetry() 
plt.grid() 
plt.title('Triangle')
plt.xlabel("Ось Х") 
plt.ylabel("Ось Y") 
plt.show() 
```

3 параллельных «ленты» из последовательностей полигонов, расположенных под острым углом к оси x; 

две пересекающихся «ленты» из последовательностей полигонов, пересекающихся не в начале координат; 

две параллельных ленты треугольников, ориентированных симметрично друг к другу; 

последовательность четырехугольников в разном масштабе, ограниченных двумя прямыми, пересекающимися в начале координат (см. рис.). 

```python
#4.1

n = int(input('Количество фигур '))
x=[]
for i in islice(count(-n), n):
    x.append(i)
x=np.array(x)
print(x)
plt.plot(x,1+x, linestyle='', marker='s', markersize=20)   
plt.plot(x,5+x, linestyle='', marker='s', markersize=20)
plt.plot(x,9+x, linestyle='', marker='s', markersize=20)
plt.grid()
plt.title( 'Rectangle')
plt.xlabel("Ось Х")
plt.ylabel("Ось Y") 
plt.show()
```
