5. Реализовать операции:
1) фильтрации фигур, являющихся выпуклыми многоугольниками
(flt_convex_polygon);
2) фильтрации фигур, имеющих хотя бы один угол, совпадающий с заданной
точкой (flt_angle_point);
3) фильтрации фигур, имеющих площадь менее заданной (flt_square);
4) фильтрации фигур, имеющих кратчайшую сторону менее заданного значе-
ния (flt_short_side);

```python
def flt_convex_polygon(polygons):
    convex_polygons = []
    for polygon in polygons:
        if polygon.convex_hull.equals(polygon):
            convex_polygons.append(polygon)
            x, y = polygon.exterior.xy
            plt.plot(x, y)
    plt.show()
    return convex_polygons

# Примеры полигонов
poly1 = Polygon([(0, 0), (1, 1), (1, 0)])
poly2 = Polygon([(0, 0), (1, 1), (2, 0), (0, 1)])
poly3 = Polygon([(0, 0), (1, 0), (1, 1), (0, 1)])

polygons = [poly1, poly2, poly3]

# Применяем функцию фильтрации
convex_polygons = flt_convex_polygon(polygons)
```

```python

def flt_angle_point(polygons, point):
    polygons_with_point = []
    for polygon in polygons:
        if point in list(polygon.exterior.coords):
            polygons_with_point.append(polygon)
            plt.figure()  # добавляем эту строку, чтобы создать новый график для каждого полигона
            x, y = polygon.exterior.xy
            plt.plot(x, y)
            plt.show()
    return polygons_with_point

# Примеры полигонов
poly1 = Polygon([(0, 0), (1, 1), (1, 0)])
poly2 = Polygon([(0, 0), (1, 1), (2, 0), (0, 1)])
poly3 = Polygon([(0, 0), (1, 0), (1, 1), (0, 1)])

polygons = [poly1, poly2, poly3]

# Применяем функцию фильтрации
polygons_with_point = flt_angle_point(polygons, (1, 1))
```

```python
from shapely.geometry import Polygon
import matplotlib.pyplot as plt

def flt_square(polygons, area):
    polygons_with_less_area = []
    for polygon in polygons:
        if polygon.area < area:
            polygons_with_less_area.append(polygon)
            plt.figure()  # добавляем эту строку, чтобы создать новый график для каждого полигона
            x, y = polygon.exterior.xy
            plt.plot(x, y)
            plt.show()
    return polygons_with_less_area

# Примеры полигонов
poly1 = Polygon([(0, 0), (1, 1), (1, 0)])
poly2 = Polygon([(0, 0), (1, 1), (2, 0), (0, 1)])
poly3 = Polygon([(0, 0), (1, 0), (1, 1), (0, 1)])

polygons = [poly1, poly2, poly3]

# Применяем функцию фильтрации
polygons_with_less_area = flt_square(polygons, 1)
```
