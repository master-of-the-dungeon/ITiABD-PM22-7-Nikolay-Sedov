```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Шаг 1: Загрузка данных
data_1 = pd.read_csv('C:\\Users\\sedof\\OneDrive\\Документы\\ITiABD-PM22-7-Nikolay-Sedov\\ML\\ML5.1 data_integration\\data\\issues1.csv', index_col=0)
data_2 = pd.read_csv('C:\\Users\\sedof\\OneDrive\\Документы\\ITiABD-PM22-7-Nikolay-Sedov\\ML\\ML5.1 data_integration\\data\\issues2.csv', index_col=0)
data_4 = pd.read_csv('C:\\Users\\sedof\\OneDrive\\Документы\\ITiABD-PM22-7-Nikolay-Sedov\\ML\\ML5.1 data_integration\\data\\issues4.csv', index_col=0)

# Шаг 2: Объединение таблиц data_1, data_2 и data_4
data_123 = pd.concat([data_1, data_2, data_4], ignore_index=True)

# Проверка на дубликаты
data_123 = data_123.drop_duplicates()

# Шаг 3: Переименование столбца в таблице orders_data
orders_data = pd.read_csv('C:\\Users\\sedof\\OneDrive\\Документы\\ITiABD-PM22-7-Nikolay-Sedov\\ML\\ML5.1 data_integration\\data\\orders.csv', index_col=0)
orders_data = orders_data.rename(columns={'Id': 'Order_id'})

# Шаг 4: Объединение таблиц data_123 и orders_data
data_with_orders = data_123.merge(orders_data, on='Order_id', how='left')

# Проверка имен столбцов в объединенной таблице
print(data_with_orders.columns)

# Шаг 5: Удаление лишних столбцов (после проверки)
columns_to_drop = [col for col in ['Unnamed: 0', 'Unnamed: 0_x', 'Unnamed: 0_y'] if col in data_with_orders.columns]
data_with_orders_cleaned = data_with_orders.drop(columns_to_drop, axis=1)

# Шаг 6: Визуализация пропусков в датасете
plt.figure(figsize=(12, 8))
sns.heatmap(data_with_orders_cleaned.isnull(), yticklabels=False, cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt





left_join = data_1.merge(data_2, on='Order_id', how='left')
print(f"Левое соединение: {left_join.shape}")

right_join = data_1.merge(data_2, on='Order_id', how='right')
print(f"Правое соединение: {right_join.shape}")


inner_join = data_1.merge(data_2, on='Order_id', how='inner')
print(f"Внутреннее соединение: {inner_join.shape}")

outer_join = data_1.merge(data_2, on='Order_id', how='outer')
print(f"Полное внешнее соединение: {outer_join.shape}")

print("Левое соединение, первая строка:")
print(left_join.iloc[0])

print("Правое соединение, первая строка:")
print(right_join.iloc[0])

print("Внутреннее соединение, первая строка:")
print(inner_join.iloc[0])

print("Полное внешнее соединение, первая строка:")
print(outer_join.iloc[0])

numerical_columns = outer_join.select_dtypes(include=['number']).columns

plt.figure(figsize=(16, 12))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(4, 4, i)
    sns.histplot(outer_join[column].dropna(), kde=True)
    plt.title(column)

plt.tight_layout()
plt.show()
```
