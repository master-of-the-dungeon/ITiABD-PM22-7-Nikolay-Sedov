```python

data_merged = pd.merge(data1, data2, on='Id', how='inner')
for column in data_merged.columns:
    if data_merged[column].dtype in ['int64', 'float64']:  
        plt.figure(figsize=(10, 4))
        sns.histplot(data_merged[column], kde=True)  
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()
```

```python
print(data1['Id'].dtype, data2['Id'].dtype)


print(data1.isnull().sum())
print(data2.isnull().sum())

data_merged = pd.merge(data1, data2, on='Id', how='inner')

print(data_merged.isnull().sum())

plt.figure(figsize=(10, 8))
sns.heatmap(data_merged.isnull(), cbar=False)
plt.title('Heatmap of Missing Values After Merging')
plt.show()

print(data_merged.sample(5))
```

```python

# Применяем разные типы соединений
inner_merged = pd.merge(data1, data2, on='Id', how='inner')
outer_merged = pd.merge(data1, data2, on='Id', how='outer')
left_merged = pd.merge(data1, data2, on='Id', how='left')
right_merged = pd.merge(data1, data2, on='Id', how='right')

# Визуализация пропущенных значений для каждого типа соединения
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8), sharex=True, sharey=True)
sns.heatmap(inner_merged.isnull(), ax=axes[0, 0], cbar=False)
axes[0, 0].set_title('Inner Join')
sns.heatmap(outer_merged.isnull(), ax=axes[0, 1], cbar=False)
axes[0, 1].set_title('Outer Join')
sns.heatmap(left_merged.isnull(), ax=axes[1, 0], cbar=False)
axes[1, 0].set_title('Left Join')
sns.heatmap(right_merged.isnull(), ax=axes[1, 1], cbar=False)
axes[1, 1].set_title('Right Join')
plt.tight_layout()
plt.show()
```
