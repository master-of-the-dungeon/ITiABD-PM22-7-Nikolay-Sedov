```python
import numpy as np
import pandas as pd
from scipy.stats import f

# Функция для преобразования строки в массив чисел
def convert(string):
    return np.array([float(num.replace(',', '.')) for num in string.split(';')])

# Исходные данные
A = "-2,63; 2,812; 3,354; -1,754; 3,732; 0,411; 4,594; 8,272; 0,437; 10,205; 5,254; -0,846; 0,729; 3,799; 1,174; 2,885; 1,489; -3,153; 2,788; 1,361; -0,556; 3,637; 0,484; -3,043; 0,471; 1,513; 4,164; 2,327; 3,715; 1,989; 3,666; 5,35; 4,509; 7,537; 0,258"
B = "0,024; -1,306; 2,986; 2,237; 3,289; 5,557; 1,174; 2,651; 3,268; 2,492; 2,516; 0,924; 5,502; 4,909; -0,901; -0,13; 0,28; -1,594; 1,728; 1,737; 1,491; 1,43; -1,218; 2,952; -0,516; 2,881; 1,591; 2,129; 0,819"
C = "-0,638; 2,623; 6,466; -1,684; 4,656; -0,655; 2,152; 1,103; 0,503; 4,185; 0,627; -0,555; -0,98; 0,852; -1,945; 5,233; 3,124; -4,2; 0,958; 4,156; 0,936; -1,321; 0,573; 0,164; 1,11; 5,099; 0,084; 0,934; 0,941; -1,156; -2,437"

# Преобразование в массивы numpy
x = convert(A)
y = convert(B)
z = convert(C)

# Размеры выборок
n1, n2, n3 = len(x), len(y), len(z)
n = [n1, n2, n3]
N = sum(n)
k = 3

# Общее среднее
grand_mean = np.mean(np.concatenate([x, y, z]))

# Межгрупповая сумма квадратов
SSTR = sum(ni * (np.mean(xi) - grand_mean) ** 2 for ni, xi in zip(n, [x, y, z]))

# Внутригрупповая сумма квадратов
SSE = sum(sum((xi - np.mean(xi)) ** 2 for xi in group) for group in [x, y, z])

# Средние квадраты
MSTR = SSTR / (k - 1)
MSE = SSE / (N - k)

# F-статистика
F = MSTR / MSE

# Уровень значимости
alpha = 0.01

# Критическое значение F
F_crit = f.ppf(1 - alpha, k - 1, N - k)

# P-значение
P = f.sf(F, k - 1, N - k)

# Вывод результатов
result = 'H0 отвергается' if F > F_crit else 'H0 не отвергается'
result, F, P
```
